```python
from google import genai
from google.colab import userdata
```

```python
api_key = userdata.get("GOOGLE_API_KEY")

client = genai.Client(api_key=api_key)

print("✅ Gemini Connected Successfully!")
```

```python
BLOG_SYSTEM_PROMPT = """
You are an experienced blog writer.

Write engaging, informative, and well-structured blogs.

The blog should include:
1. Catchy Title
2. Introduction
3. 2-3 Body Sections with headings
4. Conclusion
"""

EMAIL_SYSTEM_PROMPT = """
You are a professional email writer.

Write clear, concise, and professional emails.

The email should include:
1. Subject
2. Greeting
3. Email Body
4. Closing
"""
```

```python
def generate_blog(topic, tone, word_count):

    prompt = f"""
{BLOG_SYSTEM_PROMPT}

Topic: {topic}

Tone: {tone}

Word Count: Approximately {word_count} words.
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt,
        config={
            "temperature": 0.7
        }
    )

    return response.text
```

```python
def generate_email(recipient, purpose, tone):

    prompt = f"""
{EMAIL_SYSTEM_PROMPT}

Recipient: {recipient}

Purpose: {purpose}

Tone: {tone}
"""

    response = client.models.generate_content(
        model="gemini-2.5-flash",
        contents=prompt,
        config={
            "temperature": 0.3
        }
    )

    return response.text
```

```python
def save_to_file(filename, content):

    with open(filename, "w", encoding="utf-8") as file:
        file.write(content)
```

```python
def main():

    while True:

        print("\n========== AI CONTENT GENERATOR ==========")
        print("1. Generate Blog")
        print("2. Generate Email")
        print("3. Exit")

        choice = input("\nEnter your choice: ")

        if choice == "1":

            topic = input("Enter Blog Topic: ")

            tone = input("Enter Tone (Formal/Casual/Professional): ")

            word_count = input("Enter Word Count: ")

            blog = generate_blog(topic, tone, word_count)

            print("\n========== GENERATED BLOG ==========\n")
            print(blog)

            save_to_file("blog_output.txt", blog)

            print("\n✅ Blog saved as blog_output.txt")

        elif choice == "2":

            recipient = input("Enter Recipient: ")

            purpose = input("Enter Purpose: ")

            tone = input("Enter Tone (Formal/Casual/Professional): ")

            email = generate_email(recipient, purpose, tone)

            print("\n========== GENERATED EMAIL ==========\n")
            print(email)

            save_to_file("email_output.txt", email)

            print("\n✅ Email saved as email_output.txt")

        elif choice == "3":

            print("\nThank you for using AI Content Generator!")
            break

        else:

            print("\n❌ Invalid choice. Please try again.")

main()
```
